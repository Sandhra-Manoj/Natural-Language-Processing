### Tokenization

```python
from nltk.tokenize import sent_tokenize


text="Tokenization is the process of replacing sensitive data with unique identification symbols that retain all the essential information about the data without compromising its security."
tokenized_text=sent_tokenize(text)
print(tokenized_text)
```

```python
text="Tokenization is the process of replacing sensitive data with unique identification symbols that retain all the essential information about the data without compromising its security."
from nltk.tokenize import word_tokenize
token=word_tokenize(text)
token
```

### finding frequency distinct in the text

```python
#finding frequency distinct in the tokens
# importing freqDist library from nltk and pasing token into FreqDist
from nltk.probability import FreqDist
fdist =FreqDist(token)
fdist
```

```python
# to find the frequency of top 10 words
fdist1=fdist.most_common(10)
fdist1
```

```python
import matplotlib.pyplot as plt
fdist.plot(30,cumulative=False)
plt.show()
```

### Chunking

```python
text="We watched a fantastic movie"
token=word_tokenize(text)
tags=nltk.pos_tag(token)
reg ="NP:{<DT>?<JJ>*<NN>}"

# creating a parser
a=nltk.RegexpParser(reg)
# parsing text
result=a.parse(tags)
print(result)
```

```python
result.draw()
```
